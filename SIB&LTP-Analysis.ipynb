```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df_large = pd.read_csv('final_results/large_SIB_ratios.csv')
df_small = pd.read_csv('final_results/small_SIB_ratios.csv')
df_mid = pd.read_csv('final_results/ratios_data_no_take.csv')
df1 = pd.read_csv('final_results/ratios_data_no_take.csv')

# Combine all datasets into a single DataFrame for easier plotting
combined_df = pd.DataFrame({
    'SIB Ratio': pd.concat([df_small["SIB Ratio"], df_mid["SIB Ratio"], df_large["SIB Ratio"], df_mid["SIB Ratio"], df1["SIB+Pivot Ratio"]], ignore_index=True),
    'Category': (['Small'] * len(df_small) + 
                 ['Medium (1)'] * len(df_mid) + 
                 ['Large'] * len(df_large) + 
                 ['Medium (1)'] * len(df_mid)+ 
                 ['Medium (2)'] * len(df1))
})

# Set the plot style to white to remove grey background and gridlines
sns.set_style("white")

# Set up the figure with smaller dimensions to make it more compact
plt.figure(figsize=(6, 4))  # Adjusted to make the image more compact

# Create the box plot using seaborn with narrower width
sns.boxplot(x='Category', y='SIB Ratio', data=combined_df, palette="Set3", width=0.4)  # Adjust width here

# Add a grey dotted line between Large and SIB+LTP Ratio
plt.axvline(x=2.5, color='grey', linestyle='--', linewidth=1)  # Adjust the x position for line placement

# Add labels "SIB" and "SIB+LTP"
plt.text(1, plt.ylim()[1] + 0.005, 'SIB', ha='center', va='center', fontsize=14, color='black')
plt.text(3, plt.ylim()[1] + 0.005, 'SIB+LTP', ha='center', va='center', fontsize=14, color='black')

# Set plot labels and title
plt.ylabel('Similarity to Optimal', fontsize=16)
plt.xlabel('Categories', fontsize=16)

# Adjust label size and make the plot compact
plt.tick_params(axis='both', labelsize=14)

# Use tight_layout to ensure everything fits in the plot
plt.tight_layout()

# Save the plot
#plt.savefig('/home/ac/IPM_MPNN/accuracy.pdf', format='pdf', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df_large = pd.read_csv('final_results/large_SIB_ratios.csv')
df_small = pd.read_csv('final_results/small_SIB_ratios.csv')
df_mid = pd.read_csv('final_results/ratios_data_no_take.csv')
df1 = pd.read_csv('final_results/ratios_data_no_take.csv')

# Prepare the first dataset (Small, Medium, Large)
df_first = pd.DataFrame({
    'SIB Ratio': pd.concat([df_small["SIB Ratio"], df_mid["SIB Ratio"], df_large["SIB Ratio"]], ignore_index=True),
    'Category': (['Small'] * len(df_small) + 
                 ['Medium'] * len(df_mid) + 
                 ['Large'] * len(df_large))
})

# Prepare the second dataset (Medium (1), Medium (2))
df_second = pd.DataFrame({
    'SIB Ratio': pd.concat([df_mid["SIB Ratio"], df1["SIB+Pivot Ratio"]], ignore_index=True),
    'Category': (['SIB'] * len(df_mid) + 
                 ['LTP'] * len(df1))
})

# Set the plot style
sns.set_style("white")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(6, 4), sharey=True)  # Shared y-axis for consistency

# First subplot: Small, Medium, Large
sns.boxplot(ax=axes[0], x='Category', y='SIB Ratio', data=df_first, palette="Set3", width=0.4)
axes[0].set_title('SIB Size Comparison', fontsize=14)
axes[0].set_xlabel('Size', fontsize=12)
axes[0].set_ylabel('Similarity to Optimal', fontsize=12)

# Second subplot: Medium (1), Medium (2)
sns.boxplot(ax=axes[1], x='Category', y='SIB Ratio', data=df_second, palette="Set3", width=0.3)
axes[1].set_title('SIB-LTP comparison', fontsize=14)
axes[1].set_xlabel('Model', fontsize=12)
axes[1].set_ylabel('')  # Avoid duplicate y-axis label

# Adjust layout to fit everything nicely
plt.tight_layout()

# Save the figure
#plt.savefig('/home/ac/IPM_MPNN/accuracy.pdf', format='pdf', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df1 = pd.read_csv('final_results/ratios_data_no_take.csv')
df_large = pd.read_csv('final_results/large_SIB_ratios.csv')
df_small = pd.read_csv('final_results/small_SIB_ratios.csv')

# Prepare DataFrames for plotting
combined_ratios = pd.DataFrame({
    'Ratio': pd.concat([df1["SIB Ratio"], df1["SIB+Pivot Ratio"]], ignore_index=True),
    'Type': (['SIB Ratio'] * len(df1) + ['SIB+LTP Ratio'] * len(df1))
})

combined_df = pd.DataFrame({
    'SIB Ratio': pd.concat([df_small["SIB Ratio"], df1["SIB Ratio"], df_large["SIB Ratio"]], ignore_index=True),
    'Category': (['Small'] * len(df_small) +
                 ['Medium'] * len(df1) +
                 ['Large'] * len(df_large))
})

# Set plot style
sns.set_style("whitegrid")

# Create a figure with one combined graph
fig, ax = plt.subplots(figsize=(6, 6))  # Single plot for the combined graph

# Create the combined box plot using seaborn
sns.boxplot(x='Type', y='Ratio', data=combined_ratios, palette="Set3", width=0.5, ax=ax)

# Set plot labels and title
ax.set_title("SIB and SIB+LTP Ratios Comparison", fontsize=14)
ax.set_ylabel('Similarity to Optimal', fontsize=12)
ax.set_xlabel('Ratio Type', fontsize=12)
ax.tick_params(axis='both', labelsize=10)

# Adjust spacing and save the plot
plt.tight_layout()
#plt.savefig('/home/ac/IPM_MPNN/combined_ratios_graph.pdf', format='pdf', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
import pandas as pd

# Load the dataset
df1 = pd.read_csv('final_results/pivot_comparison.csv')

# Filter out TLE values and convert columns to numeric
filtered_df = df1[(df1['Default'] != 'TLE') & (df1['SIB'] != 'TLE')]
filtered_df['Default'] = pd.to_numeric(filtered_df['Default'], errors='coerce')
filtered_df['SIB'] = pd.to_numeric(filtered_df['SIB'], errors='coerce')

# Calculate the average decrease for cases where SIB < Default
decrease_cases = filtered_df[filtered_df['SIB'] > filtered_df['Default']]
average_decrease = ((decrease_cases['SIB'] - decrease_cases['Default']) / decrease_cases['SIB']).mean() * 100  # In percentage

print(f"Average decrease in pivots (SIB compared to Default): {average_decrease:.2f}%")
print(decrease_cases['SIB'])

```
